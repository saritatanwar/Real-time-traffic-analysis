```python
1# Distribution of Accidents by Severity
import matplotlib.pyplot as plt

# Plot the distribution of accidents by severity level
data['Severity'].value_counts().plot(kind='bar')
plt.title('Distribution of Accidents by Severity')
plt.xlabel('Severity Level')
plt.ylabel('Number of Accidents')
plt.show()
```

### Distribution of Accidents by Severity

This visualization shows the distribution of accidents by severity level. The `Severity` column categorizes accidents based on their impact, with higher levels indicating more serious incidents. By analyzing this distribution, we can see the relative frequency of minor versus severe accidents, providing insight into the overall safety conditions on the road and helping prioritize traffic safety efforts.### Observations and Insights: Distribution of Accidents by Severity

From the bar chart, we observe the frequency distribution of accidents across different severity levels. Some potential insights include:

- **Severity Levels**: The chart shows how accidents are categorized by severity, allowing us to see if most accidents are minor (low severity) or if there is a significant proportion of higher-severity incidents.
- **Most Common Severity Level**: If a particular severity level has a high count, this may suggest that many accidents are minor, possibly due to conditions like heavy traffic, where accidents occur frequently but are less severe.
- **Implications for Road Safety**: If a substantial number of accidents fall under high-severity categories, it indicates a need for targeted road safety measures. Understanding the distribution can help in planning interventions or allocating resources, such as emergency response teams or preventive safety measures.
  
By analyzing this distribution, road safety planners can prioritize areas for improvement and gain insights into the types of accidents that are most prevalent.



```python
2# Accidents by Year
# Ensure Start_Time column is parsed to datetime, coercing errors
data['Start_Time'] = pd.to_datetime(data['Start_Time'], errors='coerce')

# Extract hour from 'Start_Time' column
data['Hour'] = data['Start_Time'].dt.hour

# Plot accidents by hour, excluding rows with NaT values in 'Start_Time'
data['Hour'].value_counts().sort_index().plot(kind='bar')
plt.title('Accidents by Hour of Day')
plt.xlabel('Hour')
plt.ylabel('Number of Accidents')
plt.show()

```

### Accidents by Year
### Observations

From the bar chart, we can observe that accident frequency varies by hour of the day. Peak hours may correlate with high traffic times, such as early mornings and late afternoons. Such patterns could indicate the influence of rush hours on accident rates.



```python
#Accidents by weather
# Plot the distribution of accidents based on weather conditions
data['Weather_Condition'].value_counts().head(10).plot(kind='bar')
plt.title('Top 10 Weather Conditions for Accidents')
plt.xlabel('Weather Condition')
plt.ylabel('Number of Accidents')
plt.show()
```

### Observations and Insights: Accidents by Weather Condition

This bar chart displays the top 10 weather conditions associated with the highest number of accidents. Some key observations and insights include:

- **Most Common Conditions**: The chart highlights which weather conditions are most frequently present during accidents. If conditions like "Clear," "Rain," or "Cloudy" are prevalent, it suggests that accidents are not only limited to severe weather but can also occur in normal conditions.
- **Impact of Adverse Weather**: Certain weather conditions, such as "Rain" or "Snow," may have a noticeable number of accidents, indicating the potential hazards associated with driving in wet or icy conditions. This emphasizes the need for extra caution or safety measures during adverse weather.
- **Policy and Safety Recommendations**: Understanding weather-related accident patterns can inform road safety initiatives. For example, during rainy or snowy conditions, authorities might consider increasing road signage, reducing speed limits, or issuing weather advisories to help reduce accidents.

This analysis suggests that while adverse weather contributes to accidents, common conditions like clear or cloudy weather still see a high number of incidents, indicating that factors beyond weather, such as traffic density or driver behavior, may also play a role.
